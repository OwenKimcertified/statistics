위의 그림처럼 A 알약과 B 알약을 비교하는 실험을 한다고 가정해봅시다.

2 천명의 사람을 모집하고 각 1000 명씩 A, B 알약 실험군에 배치하고 복용했을 때 새로운 면역체계가 생성되는지를 실험해봅니다.

A 알약을 먹은 실험자들 중 990 명은 새로운 면역체계가 생성되고 10 명은 생성되지 않았고,

B 알약을 먹은 실험자들 중 120 명은 새로운 면역체계가 생성되고 880 명은 생성되지 않았습니다.

위의 상황에서 A 알약이 B 알약보다 새로운 면역체계 생성에 더 효과적이다 라는 것은 꽤나 명백해보입니다.

A 알약을 복용 후 면역체계가 생성된 사람들 중에는 플라시보 효과로 생성되었을 가능성도 있고,

B 알약을 복용 후 면역체계가 생성되지 않은 사람들 중에는 특이반응, 알러지 등으로 새로운 면역체계가 생성되지 않았을 가능성도 있죠.

하지만 명백하게 A 알약으로 새로운 면역체계를 생성한 사람이 많기 때문에,

그러한 여러가지 이유들로 A 알약이 B 알약보다 비슷하거나 약간 더 좋거나 암튼 그런거 아니야? 라고 말을 하기엔 설득력이 떨어집니다.

하지만 만약 A 알약을 먹고 150 명 중 30 명은 새로운 면역체계를 생성했고 120 명은 생성하지 못했고,

B 알약을 먹고 150 명 중 38명은 새로운 면역체계를 생성했고 112 명은 생성하지 못했다면....

우리도 항상 모든 테스트는 완벽하지 않고, 항상 무수히 많은 조건과 임의의 것들이 존재하는 상태라는 것은 인지하고 있습니다.

그 결과 우리는 확신을 가지고 B 알약이 더 좋은 알약이다 ! 라고 그 누구도 말할 수 없습니다.

이러한 상황에서 우리는 P-value 가 필요한 것입니다.

# P - value

P - value 는 0 ~ 1 사이의 값입니다. 

P - value 를 이용하면, 'A 알약이 B 알약과 차이가 있다'에 대해 얼만큼 확신할 수 있는가를 확률로 표현할 수 있습니다.

위의 상황에서 P - value 가 0 에 가까운 수라면, A 알약과 B 알약이 차이가 있다고 확신있게 말할 수 있을 것입니다.

그럼 P - value 가 __도대체 얼만큼 작아야__ 우리는 A 알약과 B 알약이 차이가 있다고 좀 더 확신할 수 있을까요?

다른 말로 좋은 결정을 내리기 위한 threshold 는 얼마일까요?? 라고 한다면, 일반적인 threshold 는 __0.05__ 입니다.

threshold 가 0.05 다 라는 의미는 __A 알약과 B 알약이 차이가 없고__, 위의 실험을 아주 많이 반복했다고 가정하면,

5 % 의 실험 결과만 틀린다는 통계적인 의미입니다.

### A 알약과 B 알약이 차이가 없고 ?? 이게 무슨 말??

A 알약을 group1, group2 에 나눠주고 결과를 확인해보니 그림과 같은 결과가 나왔고, 원인을 살펴보니 group1 에는 A약에 대한 거부 반응이 심한 사람들이 다 모여있었고, group2 에는 A약에 대한 거부 반응이 거의 없는 사람들이 다 모여있어서 였다고 해봅시다.

하지만 위 그림과 같은 결과가 나온다면 그것은 정말 각종 임의의 것들이 콜라보로 합쳐진 결과물일 것입니다.

누군가는 플라시보 효과 때문에, 누군가는 알러지 때문에, 누군가는 면역 거부 반응이라던지, 약을 먹기 전에 건강이 좋지 않았다던지..... -->

그래서 위와 같은 상황에서 P - value 를 계산해보면 0.05 보다는 작은 값이 나올 것입니다.

이럴 때 우리는 두 그룹 간 차이를 보는데 성공했다고 말합니다.

저것과 같은 실험을 반복적으로 했을 때 P - value 의 값은 0.05 보다 큰 값이 나오긴 매우 힘들 것입니다.

즉 100번을 실험하면 95번 정도는 차이가 있다고 나오고 5번 정도는 차이가 없다라고 나오겠죠.

<!-- (만약 rare 한 case 가 있다고 하면 0.05 보다 큰 값이 나올 순 있습니다. 확률적으로 가능은 하니까요)   -> 거짓양성의 case) -->

그래서 __A 알약과 B 알약이 차이가 없고 ?? 이게 무슨 말??__ 에 답을 하자면,

A 알약과 B 알약이 차이가 없다면, 5 % 의 실험 결과만이 P - value 가 0.05 보다 낮게 나올 것이야 ! 라는 것이죠.

그니까 정말 정말 rare 한 경우가 5 % 의 확률로 얻어걸려서 차이가 없다라고 말할 수도 있는 것이죠. ( 이해하기 좀 어려울 수 있음 계속보면 이해 감.)

만약 차이가 있다 라고 말하는 것이 매우매우 중요할 때(두 집단이 같은 A 알약을 먹었음에도 차이가 있는 경우를 증명) threshold 를 매우 작게 설정 할 수도 있어요(0.0.....)

그리고 만약 A 알약과 B 알약이 차이가 있다면 P - value 의 값이 0.05 보다 작을 것입니다.

# 용어

__가설검증(Hypothesis testing)__ 

H0 귀무가설 : 알약들이 차이가 없다.  < - >   H1 대립가설 : 알약들이 차이가 있다.

P - value : 귀무가설이 참이라는 가정하에 만들어짐. ★ -> A 알약과 B 알약이 차이가 있나?? (임의의 것들은 때놓고 생각해보자고 ! 의미 내포)

만약 P - value 가 0.05 보다 작다면 귀무가설을 기각하고 대립가설을 채택, 0.05 보다 크다면 대립가설을 기각하고 귀무가설을 채택. ★

그리고 P - value 의 값이 0.0000.... 이렇게 아주 작은 숫자가 나왔다고 해서 

그것이 효과의 크기(진짜 효과 좋은 알약의 탄생이다 혹은 A 알약과 B 알약의 차이가 크다)를 의미하는 것은 절대 아닙니다. 

단지 알약간 차이가 있나 없나? 만을 의미하는 것 입니다.
