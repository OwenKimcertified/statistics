# Two - sided P - value

### Discrete case

<img src = 'https://drive.google.com/uc?id=1DWH-Lp9taKsdtscKlB3BjXrA9m-LTeo_' width = 600, height = 300>

P - value 를 계산 하는 법은

1. 무작위로 관찰 시 나올 확률

2. 동일하게 희귀한 다른 관찰'들'이 나올 확률

3. 더 희귀하거나 더 극단적인 다른 관찰'들'이 나올 확률

1, 2, 3 의 확률들을 다 더하는 것입니다.

예를 들어

앞면이 두 번 연속으로 나와 '내 동전은 특별해' 라고 말을 하기 위해선 검증을 해야하고

그 검증을 P - value 로 해보는 것이죠.

귀무가설(null hypothesis) 은 '내 동전은 특별하지 않아' 이고,

대립가설(Alternative hypothesis) 은 내가 원하는 결과인 '내 동전은 특별해' 입니다.

만약 귀무가설이 참이라고 가정하고 만들어진 P - value 의 값이 0.05 보다 작다면,

정말 내 동전은 특별한 것이라고 말할 수 있겠습니다.

그래서 앞면이 두번 나왔을 때의 P - value 를 계산해보면,

.25(1.을 따름) + .25(2.을 따름 뒷면이 두번 나올 확률) + 0

P - value 가 0.5 가 나와 귀무가설을 채택해야 하는 것이죠. (그니까 별로 특별하지 않다.. 말이 그냥 복잡할 뿐)

만약 앞 1 뒤 1 이 나올 P - value 를 구한다고 하면, 0.5 가 나옵니다. -> (특별하지 않음)

앞 뒤, 뒤 앞 이 나오는 경우는 같은 경우이기 때문입니다. (독립 사건입니다. 다음 사건에 영향을 주지 않음)

<img src = 'https://drive.google.com/uc?id=1dfeyHLG3dxBdl2wU8g-WladsOHX98v_q' width = 600 height = 300>

그렇다면 다음과 같이 5 개의 동전을 던졌을 때 P - value 는 어떻게 나올까요

만약 한 개의 동전만이 뒷 면이 나올 때의 P - value 를 구한다면,

5/32(1.을 따름) + 5/32(2.을 따름 뒷4 앞1의 확률) + 2/32(3.을 따름 뒷5, 앞5의 확률합)

3/8 -> 0.375 즉 임계치인 0.05 를 상회 = 특별하지 않다. 가 됩니다. 

### Continuous case

연속적인 값들에 대한 확률, P - value 를 구할 땐 통계 분포들을 이용합니다.

예를 들어 xx 고등학교에서 여자의 키에 대한 분포가 위와 같다고 해봅시다.

곡선 아래 빨간색 빗금이 쳐진 부분들은 어떤 여학생의 키가 해당 범위에 들어갈 확률을 나타냅니다.(면적이 확률)

140 ~ 165 사이의 값들은 __곡선 아래 95 % 의 영역__ 입니다.

즉 xx 고등학교에서 여학생의 키를 물어봤을 때 __95 % 확률로 140 ~ 165 사이의 키를 가진다__ 는 뜻 입니다.

x < 140, x > 165 에 있을 확률은 각 0.025 % 입니다.

즉 xx 고등학교에서 여학생의 키를 물어봤을 때 키가 140 보다 작거나,

165 보다 클 확률이 2.5 % 다 라는 뜻 입니다.


만약 키가 138 인 학생의 P - value 를 구해본다면,

0.007(1.을 따름 : 대략적임) + 0.007(2.을 따름) + 0(3.을 따름 : 없음)

0.014 로 임계치인 0.05 보다 작기 때문에,

우리는 키가 167인 학생은 특별하다라고 생각했던 것을 그렇게 믿을 수 있는 것입니다. (대립 가설)

즉 키가 138인 학생의 키는 현재의 분포로는 설명력이 떨어지고

다른 분포에서 더 설명하기 적합하다고 말할 수 있습니다.

만약 키가 구간1 : 151 < x < 153 사이인 학생의 P - value 를 구해본다면, (140 ~ 165 를 12칸으로 쪼개고 그 중 한 구간이니까 1/12로 계산) 

0.083(1.을 따름 : 대략적임) + 0(2.을 따름 : 없음) + 

(0.4585(구간1 기준 왼쪽 부분 전부 (오른쪽보단 큰 값이여함. 편의상 계산.)) + 0.4585(오른쪽 부분 전부))(3.을 따름 : 대략적임)

P - value 는 1이 나오기 때문에 지금의 분포로 충분히 설명할 수 있고

151 < x < 153 사이인 여학생의 키는 특별하지 않았다라고 생각할 수 있습니다.(귀무가설 채택.)

# One - sided P - value

<img src = 'https://drive.google.com/uc?id=1I6gy84EhcNG64LIYYVrAYLRbpxPkyAoR' width = 600 height = 300>

단측으로 P - value 를 구할 때엔 제일 먼저 어느 방향으로 변화의 기준을 잡을 것인지 정하는 것입니다.

만약 어떤 신약을 개발하고 약의 효과가 얼마나 빨리 일어나길 본다고 했을 땐 x 값이 좌측으로 가면 갈 수록 좋은 것이겠죠.

case1 의 경우 양측으로 P -value 를 구했다면 0.05 보다 작은 값이 나와 결과가 좋게, 혹은 좋지 않게 나와도 

무언가 특수한 상황이다 라는 것을 인지할 수 있습니다.(지금의 분포로는 설명력이 낮을 정도로 특수한 상황)

하지만 단측으로 기간이 짧아지는 쪽을 본다 정하고 P - value 를 구하게되면 양측으로 했을 때의 반값으로 나오게 될 것입니다.

하지만 case2 의 경우가 나왔는데 기간이 짧아지는 쪽으로 본다 정하고 P - value 를 구하게되면 

0.9 가 넘는 값으로 나와 기존 분포로 설명할 수 있다. 

즉 '나쁘지 않군' 이라고 해석될 여지가 있습니다(매우 농후, 실제로는 별로 좋지 않은 약임에도 괜찮다고 결과가 나옴).

그래서 단측으로 P - value 를 구해야하는 때엔 분포의 내용물들에 대한 도메인 지식을 전수받을 수 있는 사람에게

많은 내용을 전달받아야 하거나 전문가에 의해서만 사용돼야 할 것입니다. (매우 직관적이지만 오해의 여지가 매우 심함.)
